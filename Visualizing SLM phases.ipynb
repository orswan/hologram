```julia
using Plots, Interact
plotly()
cd(raw"C:\Users\SrLab\desktop\swan\hologram")
push!(LOAD_PATH,pwd())
import hologram
H = hologram
```

Our trial SLM:

```julia
S = H.SLM((10,10),(4,4),(50,50),range(0,stop=2pi,length=50))
```

We'll make a phase function that is a shifted parabola:

```julia
p(x,y) = @. (x^2 + y^2)/r0^2 + x/x0
x0 = 1000;
r0 = 10;
```

Visualize the phase:

```julia
plot(H.pixelizePhase(S,p,refinement=10),st=:surface,c=:blues)
```

Look at the Fourier transform (image of the SLM):

```julia
plot(H.ft(S,p,20,refinement=10),st=:surface,c=:blues)
```

Watch how the phase and Fourier transform change with the input phase curvature:

```julia
function fout(r::Number) 
    ph(x,y) = (x.^2 + y.^2)/r.^2
    F = H.discretizePhase(S,ph,refinement=10)
    return F
end
```

```julia
mp2 = @manipulate for r in slider(.02:.02:20,label="Phase curvature")
    plot(fout(r,44))
end
```

```julia
function ftSlider(S::H.SLM,r::Number,x0::Number,y0::Number,w::Number,refinement::Int,idx::Int,returnx=true)
    phase(x,y) = (y.^2 + x.^2)/r^2 + x*x0 + y*y0
    F = H.ft(S,phase,w,refinement=refinement)
    if returnx
        return F[:,idx]
    else
        return F[idx,:]
    end
end

function dpSlider(S::H.SLM,r::Number,x0::Number,y0::Number,refinement::Int,idx::Int,returnx=true)
    phase(x,y) = (y.^2 + x.^2)/r^2 + x*x0 + y*y0
    F = H.discretizePhase(S,phase,refinement=refinement)
    if returnx
        return F[:,idx]
    else
        return F[idx,:]
    end
end

function dpSlider(S::H.SLM,r::Number,x0::Number,y0::Number,refinement::Int)
    phase(x,y) = (y.^2 + x.^2)/r^2 + x*x0 + y*y0
    F = H.discretizePhase(S,phase,refinement=refinement)
    return F
end
```

```julia
mpdp = @manipulate for r in slider(.02:.02:20,label="Phase curvature"), x0 in slider(-10:.2:10,label="X phase shift"), y0 in slider(-10:.2:10,label="Y phase shift")
    plot(dpSlider(S,r,x0,y0,10),st=:surface,c=:blues)
end
```

```julia
mpft = @manipulate for r in slider(.02:.02:20,label="Phase curvature"), x0 in slider(-10:.2:10,label="X phase shift"), y0 in slider(-10:.2:10,label="Y phase shift"), w in slider(1.0:1.0:50.0,label="Beam waist")
    plot(ftSlider(S,r,x0,y0,w,10,S.pixels[2]*5,true))
end
```
